```python
import re
import random
import sqlite3
from flask import Flask, render_template, request, redirect, url_for, session
from openai import OpenAI
import time

db_path = 'PUT PATH'
```

```python
# CORRECTNESS

tally = {}

db_column = 'correctness_gpt3_5turbo'

for i in range(1, 27):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # SQL query to count occurrences of each value in the column
    query = f"""
    SELECT {db_column}, COUNT(*) AS count
    FROM principle{i}
    GROUP BY {db_column};
    """

    # Execute the query
    cursor.execute(query)

    # Fetch all results
    results = cursor.fetchall()

    # Close the connection
    conn.close()

    for row in results:
        value, count = row

        if value in tally:
            tally[value] += count
        else:
            tally[value] = count

# Print the tally
for value, count in tally.items():
    print(f"{value}: {count}")

correct_with_principle = tally['11'] + tally['10']
print("correct_with_principle ", correct_with_principle )
correct_without_principle = tally['11'] + tally['01']
print("correct_without_principle ", correct_without_principle)
incorrect_with_principle = tally['00'] + tally['01']
print("incorrect_with_principle ", incorrect_with_principle)
incorrect_without_principle = tally['00'] + tally['10']
print("incorrect_without_principle ", incorrect_without_principle)
```
